```python
import requests
from bs4 import  BeautifulSoup
import bs4
import re
def getHTMLText(url):
    try:
        r=requests.get(url,timeout=30)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''
    
    
def getCitycode(city):
    url='http://www.360doc.com/content/14/0322/07/59625_362614659.shtml'
    html=getHTMLText(url)
    #print(html)
    try:
        list=re.findall(r'\d{9}='+str(city)+'<',html[14000:-20000])
        for i in range(len(list)):
            code=eval(list[i].split('=')[0])
    except:
        print('失败')
#    print(code)
    
    return(code)
    


def getWList(datalist,zhuangtai,maxtump,mintump,start_url):
    html=getHTMLText(start_url)
   # print(html)
    soup=BeautifulSoup(html,"html.parser")
    
    a = soup.find_all('h1')
    #print(a)
    for i in a:
        try:
            datalist.append(re.findall(r"\d{1,2}日（[\u4e00-\u9fa5]{2}）",str(i))[0])
            print(datalist[i])
        except:
            continue
   
            
    b = soup.find_all('p') 
    for i in b:
        try:
            href= i.attrs["title"]
            zhuangtai.append(re.findall(r"[\u4e00-\u9fa5]{2,4}",href)[0])
        except:
            continue
            
    c = soup.find_all('span')
    for i in c:
        try:
            maxtump.append(re.findall(r"\d{1,2}",str(i))[0])
        except:
            continue
            
    d = soup.find_all('i') 
    for i in d:
        try:
            mintump.append(re.findall(r"\d{1,2}℃",str(i))[0])
        except:
            continue
            
    
def PrintWList(datalist,zhuangtai,maxtump,mintump,day):
    
    i=int(day)-1
    d=datalist[i]
    z=zhuangtai[i]
    max1=maxtump[i]
    min1=mintump[i]
    print('{:^8}\t{:^8}\t{:^5}\t{:^5}\t'.format('日期','天气状态','最大气温','最小气温'))
    print(" ")
    print('{0:^8}\t{1:^8}\t{2:^8}\t{3:^8}\t'.format(d,z,max1,min1))
    
    
def main():
        
        while True:

            datalist=[]
            zhuangtai=[]
            maxtump=[]
            mintump=[]
            citycode='大连'
            
            while citycode!='quit':
                
                citycode=getCitycode(input('请输入你想搜索的城市(输入quit退出)：'))
                #print(citycode)

                start_url='http://www.weather.com.cn/weather/'+str(citycode)+'.shtml#'
                #print(start_url)
                getWList(datalist,zhuangtai,maxtump,mintump,start_url)
                PrintWList(datalist,zhuangtai,maxtump,mintump,input('你可以查询未来七天的天气，输入1代表今天，2代表明天，以此类推：'))
                #print(maxtump)
                
            datalist.clear()
            zhuangtai.clear()
            maxtump.clear()
            mintump.clear()

main()
```
